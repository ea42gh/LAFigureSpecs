# la_figures Binder demo

`la_figures` computes linear algebra figure specifications and delegates layout/rendering to `matrixlayout`. This notebook exercises the main interactive paths.

```python
import sympy as sym
from IPython.display import SVG, display

import la_figures as lf

RENDER_OPTS = dict(
    toolchain_name="pdftex_dvisvgm",
    crop="tight",
    padding=(2, 2, 2, 2),
    output_dir="/tmp/la/run",
)

def show(svg):
    display(SVG(svg))
```

## Row reduction of an augmented system

```python
A = sym.Matrix([[1, 2], [1, 2], [3, 4]])
spec = lf.ge_tbl_spec(A, gj=False)
spec["outer_hspace_mm"] = 8
spec["cell_align"] = "r"

svg = lf.render_ge_svg(**spec, **RENDER_OPTS)
show(svg)
```

## Gaussian elimination with a right-hand side

```python
A = sym.Matrix([[-2, 0, -2, 2, 0], [2, 0, 3, -3, 0], [0, 0, 0, 0, 1]])
b = sym.Matrix([4, -5, -1])

svg = lf.ge_tbl_svg(A, ref_rhs=b, gj=False, fig_scale=1.1, **RENDER_OPTS)
show(svg)
```

## Back-substitution fragments

```python
U = sym.Matrix([[1, 2, 0], [0, 1, 3]])
rhs = sym.Matrix([4, 5])

preamble = r"""
\usepackage{amsmath,amssymb,mathtools,xcolor}
\usepackage{systeme}
\usepackage{nicematrix}
\usepackage{cascade}
"""
cascade_tex = "\n".join(lf.backsubstitution_tex(U, rhs))
print(cascade_tex)

show(lf.latex_svg("$" + lf.linear_system_tex(U, rhs) + "$", preamble=preamble, scale=0.9, **RENDER_OPTS))
show(lf.latex_svg(lf.standard_solution_tex(U, rhs), preamble=preamble, scale=0.9, **RENDER_OPTS))
```

## Gram-Schmidt / QR layout

```python
A = sym.Matrix([[-1, -4, -3], [1, 0, -3], [1, 4, 1], [-1, 0, 1]])

svg = lf.gram_schmidt_qr(A, fig_scale=1.1, **RENDER_OPTS)
show(svg)
```

## Eigenproblem table

```python
A = sym.Matrix([[2, 1], [0, 3]])

svg = lf.eig_tbl_svg(A, fig_scale=1.1, **RENDER_OPTS)
show(svg)
```

## SVD table

```python
A = sym.Matrix([
    [sym.Rational(6, 35), 0, sym.Rational(-8, 35), sym.Rational(3, 7)],
    [sym.Rational(18, 35), 0, sym.Rational(-24, 35), sym.Rational(2, 7)],
    [sym.Rational(9, 35), 0, sym.Rational(-12, 35), sym.Rational(-6, 7)],
])

svg = lf.svd_tbl_svg(A, fig_scale=1.1, **RENDER_OPTS)
show(svg)
```
