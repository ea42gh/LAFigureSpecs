# QR Top-Level Functions

This notebook groups QR/Gram-Schmidt top-level `la_figures` functions by task.


```python
import sympy as sym
from IPython.display import SVG, display
import la_figures as lf
```

## 1) Compute QR matrix stacks

- `compute_qr_matrices(A, W)`
- `gram_schmidt_qr_matrices(A, W, ...)`


```python
A = sym.Matrix([[1, 2], [3, 4], [5, 6]])
W = sym.Matrix([[1, 0], [0, 1], [0, 0]])

mats = lf.compute_qr_matrices(A, W)
print(len(mats), len(mats[0]))

gs_mats = lf.gram_schmidt_qr_matrices(A, W)
print(len(gs_mats), len(gs_mats[0]))
```

## 2) Build specs

- `qr_tbl_spec(A, W, ...)`
- `qr_tbl_layout_spec(A, W, ...)`
- `qr_matrices_from_grid(mats)`


```python
spec = lf.qr_tbl_spec(A, W)
layout_spec = lf.qr_tbl_layout_spec(A, W)
A0, W0, WtA, WtW, S, Qt, Q, R = lf.qr_matrices_from_grid(mats)
print(spec.keys())
print(layout_spec.keys())
print(R.shape)
```

## 3) Bundle / TeX / SVG wrappers

- `qr_tbl_bundle(A, W, ...)`
- `qr_tbl_tex(A, W, ...)`
- `qr_tbl_svg(A, W, ...)`


```python
bundle = lf.qr_tbl_bundle(A, W)
print(bundle.keys())

tex = lf.qr_tbl_tex(A, W)
svg = lf.qr_tbl_svg(A, W)
display(SVG(svg))
```

## 4) Render-only / compute+render helpers

- `qr(matrices, ...)`: render from a precomputed stack
- `gram_schmidt_qr(A, W, ...)`: compute+render


```python
svg2 = lf.qr(mats)
svg3 = lf.gram_schmidt_qr(A, W)
display(SVG(svg2))
```
