# QR Top-Level Functions

This notebook groups QR/Gram-Schmidt top-level `LAFigureSpecs` functions by task.


```python
import sympy as sym
from IPython.display import SVG, display
import LAFigureSpecs as lf
```

## 1) Compute QR matrix stacks

- `compute_qr_matrices(A)`
- `qr_tbl_spec(A, ...)`


```python
A = sym.Matrix([[1, 2], [3, 4], [5, 6]])

mats = lf.compute_qr_matrices(A)
print(len(mats), len(mats[0]))

gs_mats = lf.qr_tbl_spec(A)["matrices"]
print(len(gs_mats), len(gs_mats[0]))
```

## 2) Build specs

- `qr_tbl_spec(A, ...)`
- `qr_tbl_layout_spec(A, ...)`
- `qr_matrices_from_grid(mats)`


Note: `qr_tbl_spec` returns a plain dict; `qr_tbl_layout_spec` returns a `QRGridSpec` object.
Use `layout_spec.to_dict()` if you need dict-style access.


```python
spec = lf.qr_tbl_spec(A)
layout_spec = lf.qr_tbl_layout_spec(A)
qrA = lf.qr_matrices_from_grid(mats)
print(spec.keys())
print(layout_spec.to_dict().keys())
print(qrA.keys())
print("size of R: ", qrA["R"].shape)
```

## 3) Bundle / TeX / SVG wrappers

- `qr_bundle(A, ...)`
- `qr_tbl_tex(A, ...)`
- `qr_tbl_svg(A, ...)`


```python
bundle = lf.qr_bundle(A)
print(bundle.keys())

tex = lf.qr_tbl_tex(A)
svg = lf.qr_tbl_svg(A)
display(SVG(svg))
```

## 4) Render-only / compute+render helpers

- `qr(matrices, ...)`: render from a precomputed stack
- `qr_figure(A, ...)`: compute+render


```python
svg2 = lf.qr(mats)
svg3 = lf.qr_figure(A)
display(SVG(svg2))
```
