# Generic LaTeX and LAlatex SVG Rendering

This notebook demonstrates the generic rendering helpers:

- `latex_svg(...)` for standalone LaTeX fragments
- `latex_document_svg(...)` for full LaTeX documents
- `lshow_svg(...)` for Julia `LAlatex.L_show(...)` output via `juliacall`


```python
from IPython.display import SVG, display
import LAFigureSpecs as lf
```

## Render a LaTeX fragment

Use `latex_svg(...)` when you already have a LaTeX fragment and want SVG output with the same render controls used by the table renderers.

The SVG wrappers resolve `output_dir`/`tmp_dir` once before rendering; use `output_dir` for new code and keep `tmp_dir` only for compatibility.

By default, fragments are wrapped with the packages used by LAFigureSpecs helpers: `amsmath`, `amssymb`, `mathtools`, `xcolor`, `systeme`, `cascade`, and `nicematrix`.


```python
svg = lf.latex_svg(
    r"$A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$",
    crop="tight",
    padding=(2, 2, 2, 2),
)
display(SVG(svg))
```

## Package-backed fragments

The defaults cover LAFigureSpecs-generated fragments such as `\systeme`, cascade steps, and `nicematrix` tables. Use `tex_packages="..."` when a fragment needs a different package set.


```python
svg = lf.latex_svg(
    r"$\systeme{x + 2y = 5, y = 1}$",
    crop="tight",
    padding=(2, 2, 2, 2),
)
display(SVG(svg))
```

`preamble=...` is an override path. If you pass it, `tex_packages`, `tikz_libraries`, and the default package list are not used, so the preamble must include every package the fragment needs.

Some LaTeX environments are sensitive to being wrapped in `\scalebox{...}{...}`. For example, `tikz-cd` should use `ampersand replacement=\&` instead of raw `&` alignment markers. In practice, `tikz-cd` is more robust through `latex_document_svg(...)` than `latex_svg(...)`.


```python
tikzcd_tex = r"""
\begin{tikzcd}[column sep=large, row sep=large, ampersand replacement=\&]
\mathbb{R}^n \arrow[r, "T"] \arrow[d, "P"'] \&
\mathbb{R}^m \arrow[d, "Q"] \\
\mathbb{R}^n \arrow[r, "A_T"'] \&
\mathbb{R}^m
\end{tikzcd}
"""

tikzcd_doc = rf"""\documentclass{{standalone}}
\usepackage{{amsmath,amssymb,tikz-cd}}
\begin{{document}}
{tikzcd_tex}
\end{{document}}
"""

svg = lf.latex_document_svg(
    tikzcd_doc,
    toolchain_name="pdftex_pdftocairo",
    crop="tight",
    padding=(2, 2, 2, 2),
)
display(SVG(svg))
```

## Render a full LaTeX document

Use `latex_document_svg(...)` when the input is already a complete document.


```python
doc = r"""
\documentclass{standalone}
\usepackage{amsmath}
\begin{document}
$x^2 + y^2 = 1$
\end{document}
"""

svg = lf.latex_document_svg(doc, crop="tight", padding=(2, 2, 2, 2))
display(SVG(svg))
```

## Bridge Julia `LAlatex.L_show(...)`

If `juliacall` is installed and the Julia package `LAlatex` is available, `lshow_svg(...)` calls Julia `LAlatex.L_show(...)` and renders the result to SVG.

Simple numeric Python vectors and matrices are normalized before the Julia call, so NumPy arrays and rectangular nested lists work naturally.


```python
svg = lf.lshow_svg(
    "A = ",
    [[1, 2], [3, 4]],
    crop="tight",
    padding=(2, 2, 2, 2),
)
display(SVG(svg))
```
