# Generic LaTeX and LAlatex SVG Rendering

This notebook demonstrates the generic rendering helpers:

- `latex_svg(...)` for standalone LaTeX fragments
- `latex_document_svg(...)` for full LaTeX documents
- `lshow_svg(...)` for Julia `LAlatex.L_show(...)` output via `juliacall`


```python
from IPython.display import SVG, display
import la_figures as lf
```

## Render a LaTeX fragment

Use `latex_svg(...)` when you already have a LaTeX fragment and want SVG output with the same render controls used by the table renderers.


```python
svg = lf.latex_svg(
    r"$A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$",
    crop="tight",
    padding=(2, 2, 2, 2),
)
display(SVG(svg))
```

## Render a full LaTeX document

Use `latex_document_svg(...)` when the input is already a complete document.


```python
doc = r"""
\documentclass{standalone}
\usepackage{amsmath}
\begin{document}
$x^2 + y^2 = 1$
\end{document}
"""

svg = lf.latex_document_svg(doc, crop="tight", padding=(2, 2, 2, 2))
display(SVG(svg))
```

## Bridge Julia `LAlatex.L_show(...)`

If `juliacall` is installed and the Julia package `LAlatex` is available, `lshow_svg(...)` calls Julia `LAlatex.L_show(...)` and renders the result to SVG.

Simple numeric Python vectors and matrices are normalized before the Julia call, so NumPy arrays and rectangular nested lists work naturally.


```python
svg = lf.lshow_svg(
    "A = ",
    [[1, 2], [3, 4]],
    crop="tight",
    padding=(2, 2, 2, 2),
)
display(SVG(svg))
```
