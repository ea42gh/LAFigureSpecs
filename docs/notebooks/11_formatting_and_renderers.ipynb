# Formatting Helpers and Renderer Entrypoints

This notebook covers top-level formatting helpers and low-level render re-exports.


```python
import sympy as sym
import LAFigureSpecs as lf
```

## Formatting helpers

- `make_decorator`, `decorate_with`, `compose_decorators`
- selectors: `sel_entry`, `sel_row`, `sel_col`, `sel_box`, `sel_diag`, `sel_antidiag`, `sel_mask`


```python
dec = lf.make_decorator(text_color="red", bf=True)
print(dec("x"))
```

## Low-level renderers (re-exported from matrixlayout)

- `render_ge_tex/svg`
- `render_qr_tex/svg`
- `render_eig_tex/svg`

Use these when you already have full specs and want direct rendering control.


```python
A = sym.Matrix([[1, 2], [3, 4]])
spec = lf.ge_tbl_spec(A)
tex = lf.render_ge_tex(**spec)
svg = lf.render_ge_svg(**spec)
print(type(tex), type(svg))
```
