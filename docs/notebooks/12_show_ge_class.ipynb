# ShowGE (stateful GE helper)

This notebook demonstrates the `ShowGE` class, which mirrors the Julia `ShowGE` workflow.
It now also highlights the direct matrix accessors `lhs_matrix`, `rhs_matrix`, and `rhs_column`.


```python
import sympy as sym
import LAFigureSpecs as lf

A = sym.Matrix([[1, 2], [3, 4]])
b = sym.Matrix([[5], [6]])

show = lf.ShowGE(A, b, gj=True)
show.ref(pivoting="partial")
show.show_layout()
```

```python
# Access computed results and extracted matrices
display(show.matrices())
display(show.lhs_matrix())
display(show.rhs_matrix())
display(show.rhs_column())
display(show.particular_solution())
display(show.homogeneous_solution())
```

```python
# System, backsubstitution, and solution blocks
show.show_system()
show.show_backsubstitution()
show.show_solution()
```

## Inconsistent systems

- `ShowGE` tracks per-RHS status in `show.rhs_status`.
- Inconsistent RHS columns are marked with a red **×** in the variable-summary row.
- `show_backsubstitution()` renders `0 = rhs` with **No Solution**.
- `show_solution()` returns an empty list for inconsistent RHS columns.


```python
# Access the coefficient and RHS blocks from the final GE stack
lhs = show.lhs_matrix()
rhs = show.rhs_matrix()
display(lhs)
display(rhs)
```

```python
# Single RHS column (0-based index)
rhs_col = show.rhs_column(0)
display(rhs_col)
```
