# Eigen/SVD Top-Level Functions

This notebook groups eigen and SVD top-level `la_figures` functions by task.


```python
import sympy as sym
from IPython.display import SVG, display
import la_figures as lf
```

## 1) Eigen helpers

- `eig_tbl_spec(A, ...)`
- `eig_matrices_from_spec(spec, orthonormal=True)`
- `eig_tbl_tex/svg/bundle(...)`


```python
A = sym.Matrix([[4, 2], [0, 9]])

espec = lf.eig_tbl_spec(A)
Lam, V = lf.eig_matrices_from_spec(espec)
print(espec.keys())
print(Lam.shape, V.shape)

ebundle = lf.eig_tbl_bundle(A)
print(ebundle.keys())

esvg = lf.eig_tbl_svg(A)
display(SVG(esvg))
```

## 2) SVD helpers

- `svd_tbl_spec(A, ...)`
- `svd_tbl_spec_from_right_singular_vectors(V, ...)`
- `svd_matrices_from_spec(spec, reduced=True)`
- `svd_tbl_tex/svg/bundle(...)`


```python
sspec = lf.svd_tbl_spec(A)
U, S, Vt, rank = lf.svd_matrices_from_spec(sspec, reduced=True)
print(sspec.keys())
print(U.shape, S.shape, Vt.shape, rank)

sbundle = lf.svd_tbl_bundle(A)
print(sbundle.keys())

ssvg = lf.svd_tbl_svg(A)
display(SVG(ssvg))
```
