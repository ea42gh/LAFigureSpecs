# Backsubstitution Top-Level Functions

This notebook covers the system/backsub text helpers in `la_figures`.


```python
import sympy as sym
from IPython.display import SVG, display
import la_figures as lf
from matrixlayout.backsubst import backsubst_svg
```

## Functions

- `linear_system_tex(A, b, ...)`
- `backsubstitution_tex(A, b, ...)`
- `standard_solution_tex(A, b, ...)`


```python
A = sym.Matrix([[1, 2, 0], [0, 1, 3]])
b = sym.Matrix([[4], [5]])

system_txt = lf.linear_system_tex(A, b)
cascade_txt = lf.backsubstitution_tex(A, b)
solution_txt = lf.standard_solution_tex(A, b)

print(system_txt)
print(solution_txt)

svg = backsubst_svg(system_txt=system_txt, cascade_txt=cascade_txt, solution_txt=solution_txt)
display(SVG(svg))
```
